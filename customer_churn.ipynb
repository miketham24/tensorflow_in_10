Import data:

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/Churn.csv')
```

```python
# Convert categorical variables to dummy variables for ML algs
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))
# Convert the target variable to binary. If churn, 1, else 0
y = df['Churn'].apply(lambda x:1 if x=='Yes' else 0)

# Split the data, 80% training, 20% testing 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train.head()
```

Import Dependencies:

```python
# Sequential (core model class), load_model allows us to load model from memory
from tensorflow.keras.models import Sequential, load_model
# Dense is a layer itself, allows us to build hidden layers in neural network
from tensorflow.keras.layers import Dense
# accuracy is metric we are basing our model off of
from sklearn.metrics import accuracy_score
```

Build and compile model:

```python
model = Sequential()
#building hidden layers
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=64, activation='relu'))
#final layer dictates what output will look like
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

Fit, Predict, and evaluate:

```python
#epochs is number of times model will go through data, training longer may improve accuracy but might lead to overfitting
model.fit(X_train, y_train, epochs=1000, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```

Saving and Reloading:

```python
model.save('tfmodel.keras')
model.export('tfmodel')
```
